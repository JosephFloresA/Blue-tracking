```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Cargar imagen
img_path = "map1_edit.png"
img = cv2.imread(img_path)

# Convertir a espacio de color HSV para segmentación por color
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Rango para color magenta
lower_magenta = np.array([140, 50, 50])
upper_magenta = np.array([170, 255, 255])

# Rango para color verde
lower_green = np.array([40, 50, 50])
upper_green = np.array([80, 255, 255])

# Rango para color gris (difícil, usar valores bajos de saturación)
lower_gray = np.array([0, 0, 100])
upper_gray = np.array([180, 30, 150])

# Crear máscaras
mask_magenta = cv2.inRange(hsv, lower_magenta, upper_magenta)
mask_green = cv2.inRange(hsv, lower_green, upper_green)
mask_gray = cv2.inRange(hsv, lower_gray, upper_gray)

# Encontrar contornos para cada color
def get_centroids(mask):
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    centroids = []
    for cnt in contours:
        M = cv2.moments(cnt)
        if M["m00"] != 0:
            cx = int(M["m10"]/M["m00"])
            cy = int(M["m01"]/M["m00"])
            centroids.append((cx, cy))
    return centroids

pts_magenta = get_centroids(mask_magenta)
pts_green = get_centroids(mask_green)
pts_gray = get_centroids(mask_gray)

(len(pts_magenta), len(pts_green), len(pts_gray))
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# Magenta
if pts_magenta:
    x_magenta, y_magenta = zip(*pts_magenta)
    plt.scatter(x_magenta, y_magenta, c='magenta', label='Magenta', s=40)

# Verde
if pts_green:
    x_green, y_green = zip(*pts_green)
    plt.scatter(x_green, y_green, c='limegreen', label='Verde', s=40)

# Gris
if pts_gray:
    x_gray, y_gray = zip(*pts_gray)
    plt.scatter(x_gray, y_gray, c='gray', label='Gris', s=40)

plt.legend()
plt.axis('off')
plt.title('Puntos detectados sobre la imagen')
plt.show()
```

```python
#cordenadas absoltas imagen
superior_izquierda = (84.6, 43)  # (°W, °N)
inferior_derecha = (45, 23)  # (°W, °N)
```

```python
# -----------------------------------------------------------
# Conversión de píxeles a coordenadas geográficas
# -----------------------------------------------------------

# Dimensiones de la imagen
h, w, _ = img.shape

# Límites geográficos del mapa (en grados)
lon_min, lon_max = -84.6, -45.0   # Oeste a Este
lat_max, lat_min = 43.0, 23.0     # Norte a Sur

def pixel_to_coords(cx, cy):
    """
    Convierte coordenadas de píxeles (cx, cy) a coordenadas geográficas (lon, lat)
    según los límites especificados del mapa.
    """
    lon = lon_min + (cx / w) * (lon_max - lon_min)
    lat = lat_max - (cy / h) * (lat_max - lat_min)  # invertido porque Y crece hacia abajo
    return lon, lat


# -----------------------------------------------------------
# Generar listas con las coordenadas de cada color
# -----------------------------------------------------------

colores, lons, lats = [], [], []

# Puntos magenta
for (cx, cy) in pts_magenta:
    lon, lat = pixel_to_coords(cx, cy)
    colores.append("Magenta")
    lons.append(round(lon, 2))
    lats.append(round(lat, 2))

# Puntos verdes
for (cx, cy) in pts_green:
    lon, lat = pixel_to_coords(cx, cy)
    colores.append("Verde")
    lons.append(round(lon, 2))
    lats.append(round(lat, 2))

# Puntos grises
for (cx, cy) in pts_gray:
    lon, lat = pixel_to_coords(cx, cy)
    colores.append("Gris")
    lons.append(round(lon, 2))
    lats.append(round(lat, 2))


# -----------------------------------------------------------
# Crear diccionario final
# -----------------------------------------------------------

datos_tiburones = {
    "color_punto": colores,
    "longitud": lons,
    "latitud": lats
}

print(f"Total de puntos: {len(colores)}")
print({k: v[:5] for k, v in datos_tiburones.items()})  # muestra ejemplo
```
